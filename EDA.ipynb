# Bike Details Datase

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("D:/k study/studyrelated things/assignments-/data/BIKE_DETAILS.csv")
```

```python
df
```

```python
#Q1 What is the range of selling prices in the datasetD ?
min = df['selling_price'].min()
max = df['selling_price'].max()
range = max - min
print(range)
```

```python
# Q2 What is the median selling price for bikes in the datasetD?
median_selling_price = df['selling_price'].median()
print(median_selling_price)
```

```python
# Q3 What is the most common seller typeD?
most_common_seller_type_ID = df['seller_type'].mode()[0]
print(most_common_seller_type_ID)
```

```python
# Q4 How many bikes have driven more than 50,000 kilometersD?
bikes_driven_more_than_50000 = len(df[df['km_driven'] > 500000])
print(bikes_driven_more_than_50000)
```

```python
# Q5 What is the average km_driven value for each ownership typeD?
avg_km_per_owner = df.groupby("owner")["km_driven"].mean()
print(avg_km_per_owner)
```

```python
# Q6  What proportion of bikes are from the year 2015 or older?
total_bikes = len(df)
older_bikes = len(df[df["year"] <= 2015])
proportion = older_bikes / total_bikes
print(proportion)
```

```python
#  Q7 What is the trend of missing values across the datasetD?
missing_values = df.isnull().sum()

missing_percentage = (missing_values / len(df)) * 100

missing_summary = pd.DataFrame({
    "Missing Values": missing_values,
    "Percentage Missing": missing_percentage
}).sort_values(by="Missing Values", ascending=False)

print(missing_summary)
```

```python
# Q8 What is the highest ex_showroom_price recorded, and for which bikeD?
max_price_row = df.loc[df['ex_showroom_price'].idxmax(), ['name', 'ex_showroom_price']]
print("Bike with the highest ex_showroom_price:")
print(max_price_row)
```

```python
# Q9 What is the total number of bikes listed by each seller typeD?
bikes_per_seller_type = df['seller_type'].value_counts()
print("Total Bikes Listed by Each Seller Type:")
print(bikes_per_seller_type)
```

```python
df.columns
```

```python
# Q10 What is the relationship between selling_price and km_driven for first-owner bikesD?
sns.scatterplot(data=df[df['owner'] == '1st owner'], x='km_driven', y='selling_price')
plt.title("Selling Price vs. KM Driven for First Owner Bikes")
plt.show()
```

```python
# Q11 Identify and remove outliers in the km_driven column using the IQR methodB.
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1
filtered_df = df[~((df['km_driven'] < (Q1 - 1.5 * IQR)) | (df['km_driven'] > (Q3 + 1.5 * IQR)))]
print("Outliers in km_driven removed using IQR method.")
```

```python
# Q12 Perform a bivariate analysis to visualize the relationship between year and selling_priceB.
sns.boxplot(data=df, x='year', y='selling_price')
plt.xticks(rotation=45)
plt.title("Bivariate Analysis: Year vs. Selling Price")
plt.show()
```

```python
# Q13  What is the average depreciation in selling price based on the bike's age (current year - manufacturingyear) ?
current_year = 2025
df['age'] = current_year - df['year']
depreciation = df.groupby('age')['selling_price'].mean()
print("Average Depreciation in Selling Price Based on Bike's Age:")
print(depreciation)
```

```python
# Q14 Which bike names are priced significantly above the average price for their manufacturing yearD?
avg_price_by_year = df.groupby('year')['selling_price'].mean()
above_avg_bikes = df[df.apply(lambda row: row['selling_price'] > avg_price_by_year[row['year']], axis=1)]
print("Bikes Priced Significantly Above Average for Their Manufacturing Year:")
print(above_avg_bikes[['name', 'year', 'selling_price']])
```

```python
# Q15  Develop a correlation matrix for numeric columns and visualize it using a heatmap.
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

# Car Sale Dataset

```python
df2 = pd.read_csv('D:/k study/studyrelated things/assignments-/data/Car_Sale.csv')
```

```python
df2.columns
```

```python
df2.head()
```

```python
# Q1 What is the average selling price of cars for each dealer, and how does it compare across different dealers?
df2['Price ($)'] = pd.to_numeric(df2['Price ($)'], errors='coerce')

dealer_avg_price = df2.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending=False)
dealer_avg_price
```

```python
# Q2 Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends?
company_price_variation = df2.groupby('Company')['Price ($)'].std().sort_values(ascending=False)
company_price_variation
```

```python
# Q3 What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?
transmission_price_distribution = df2.groupby('Transmission')['Price ($)'].describe()[['25%', '50%', '75%']]
transmission_price_distribution
```

```python
# Q4  What is the distribution of car prices across different regions?
region_price_distribution = df2.groupby('Dealer_Region')['Price ($)'].describe()[['25%', '50%', '75%']]
region_price_distribution
```

```python
# Q5 What is the distribution of cars based on body styles?
body_style_distribution = df2['Body Style'].value_counts()
body_style_distribution
```

```python
# Q6 How does the average selling price of cars vary by customer gender and annual income?
gender_price_distribution = df2.groupby('Gender')['Price ($)'].mean()
income_brackets = pd.qcut(df2['Annual Income'], q=4, labels=["Low", "Medium", "High", "Very High"])
income_price_distribution = df2.groupby(income_brackets)['Price ($)'].mean()
```

```python
# Q7  What is the distribution of car prices by region, and how does the number of cars sold vary by region?
region_car_count = df2['Dealer_Region'].value_counts()
region_car_count
```

```python
# Q8 How does the average car price differ between cars with different engine sizes?
engine_avg_price = df2.groupby('Engine')['Price ($)'].mean().sort_values(ascending=False)
engine_avg_price
```

```python
# Q9  How do car prices vary based on the customerâ€™s annual income bracket?
income_price_variation = df2.groupby(income_brackets)['Price ($)'].std()
income_price_variation
```

```python
# Q10 What are the top 5 car models with the highest number of sales, and how does their price distribution look
top_5_models = df2['Model'].value_counts().head(5).index
top_5_models_price_distribution = df2[df2['Model'].isin(top_5_models)].groupby('Model')['Price ($)'].describe()
top_5_models_price_distribution
```

```python
# Q11 How does car price vary with engine size across different car colors, and which colors have the highest price variation?
color_price_variation = df2.groupby('Color')['Price ($)'].std()
engine_color_price_distribution = df2.groupby(['Engine', 'Color'])['Price ($)'].mean().unstack()
```

```python
# Q12 Is there any seasonal trend in car sales based on the date of sale?
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')
df2['Month'] = df2['Date'].dt.month
monthly_sales = df2['Month'].value_counts().sort_index()
monthly_sales
```

```python
# Q13 How does the car price distribution change when considering different combinations of body style and transmission typ?
body_trans_price_distribution = df2.groupby(['Body Style', 'Transmission'])['Price ($)'].describe()[['25%', '50%', '75%']]
body_style_distribution
```

```python
# Q14 What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?
df2_numeric = df2[['Price ($)', 'Annual Income']].dropna()
correlation_matrix = df2_numeric.corr()
correlation_matrix
```

```python
# Q15 How does the average car price vary across different car models and engine types?
model_engine_price = df2.groupby(['Model', 'Engine'])['Price ($)'].mean().unstack()
model_engine_price
```

```python
df3['discounted_price'] = df3['discounted_price'].astype(str).str.replace('â‚¹', '').str.replace(',', '').astype(float)
df3['actual_price'] = df3['actual_price'].astype(str).str.replace('â‚¹', '').str.replace(',', '').astype(float)
df3['discount_percentage'] = df3['discount_percentage'].astype(str).str.replace('%', '').astype(float)
df3['rating'] = pd.to_numeric(df3['rating'], errors='coerce')
df3['rating_count'] = df3['rating_count'].astype(str).str.replace(',', '').astype(float)
```

```python
# Q1 What is the average rating for each product category?
avg_rating_by_category = df3.groupby('category')['rating'].mean().sort_values(ascending=False)
avg_rating_by_category
```

```python
# Q2 What are the top rating_count products by category?
top_rating_count_by_category = df3.loc[df3.groupby('category')['rating_count'].idxmax(), ['category', 'product_name', 'rating_count']]
top_rating_count_by_category
```

```python
# Q3 What is the distribution of discounted prices vs. actual price?
plt.figure(figsize=(10, 12))
sns.histplot(df3[['discounted_price', 'actual_price']], kde=True, bins=30)
plt.title("Distribution of Discounted vs. Actual Prices")
plt.xlabel("Price")
plt.legend(["Discounted Price", "Actual Price"])
plt.show()
```

```python
# Q7 What are the most popular product review?
top_reviews = df3[['review_content', 'rating_count']].nlargest(10, 'rating_count')
top_reviews
```

```python
# Q8 What is the correlation between discounted_price and rating?
correlation_price_rating = df3[['discounted_price', 'rating']].corr().iloc[0, 1]
correlation_price_rating
```

```python
# Q9  What are the Top 5 categories based on the highest rating?\
top_categories_by_rating = avg_rating_by_category.head(5)
top_categories_by_rating
```

# Spotify DataSet

```python
df4 = pd.read_csv('D:/k study/studyrelated things/assignments-/data/spotify.csv')
```

```python
df4.head()
```

```python
df4.shape
```

```python
# Q1 Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful?
print("Missing Values:\n", df4.isnull().sum())
print("Duplicate Rows Before Removal:", df4.duplicated().sum())
df4 = df4.drop_duplicates()
print("Duplicate Rows After Removal:", df4.duplicated().sum())
```

```python
# Q2 What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram?
plt.figure(figsize=(8, 5))
sns.histplot(df4["Popularity"], bins=20, kde=True, color="blue")
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()
```

```python
# Q3  Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot?
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df4["Duration (ms)"], y=df4["Popularity"], alpha=0.6, color="red")
plt.title("Relationship Between Popularity and Duration of Tracks")
plt.xlabel("Duration (ms)")
plt.ylabel("Popularity")
plt.show()
```

```python
# Q4 Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot?
artist_counts = df4['Artist'].value_counts()
plt.figure(figsize=(10, 50))
sns.countplot(y=artist_counts.index, order=artist_counts.index, palette="viridis")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.title("Number of Tracks per Artist")
plt.show()
```

```python
# Q5 What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each?
least_popular_tracks = df4.nsmallest(5, "Popularity")[["Artist", "Track Name", "Popularity"]]
print("Top 5 Least Popular Tracks:\n", least_popular_tracks)
```

```python
# Q6 Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist?
top_5_artists = df4.groupby("Artist")["Popularity"].mean().nlargest(5)
print("Top 5 Most Popular Artists by Average Popularity:\n", top_5_artists)
```

```python
# Q7 For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist?
most_popular_tracks = df4.loc[df4.groupby("Artist")["Popularity"].idxmax(), ["Artist", "Track Name", "Popularity"]]
most_popular_tracks = most_popular_tracks[most_popular_tracks["Artist"].isin(top_5_artists.index)]
print("Most Popular Tracks of the Top 5 Artists:\n", most_popular_tracks)
```

```python
# Q8 Visualize relationships between multiple numerical variables simultaneously using a pair plot.
sns.pairplot(df4[["Popularity", "Duration (ms)"]])
plt.show()
```

```python
# Q9 Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.
plt.figure(figsize=(12, 6))
top_10_artists = df4["Artist"].value_counts().index[:10]
sns.violinplot(x=df4[df4["Artist"].isin(top_10_artists)]["Artist"],y=df4[df4["Artist"].isin(top_10_artists)]["Duration (ms)"], palette="coolwarm")
plt.xticks(rotation=45, ha='right')
plt.title("Variation in Track Duration Across Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.show()
```

```python
# Q10  How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot?
plt.figure(figsize=(12, 6))
sns.swarmplot(x=df4[df4["Artist"].isin(top_10_artists)]["Artist"], y=df4[df4["Artist"].isin(top_10_artists)]["Popularity"], palette="muted")
plt.xticks(rotation=45, ha='right')
plt.title("Distribution of Track Popularity Across Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Popularity")
plt.show()
```
